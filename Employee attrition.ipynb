```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
```

```python
df= pd.read_csv(r"C:\Users\ASUS\Downloads\Attrition data.csv")
```

```python
replaced_columns = ['PercentSalaryHike', 'StandardHours', 'TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','EnvironmentSatisfaction','YearsSinceLastPromotion','JobSatisfaction','WorkLifeBalance','JobInvolvement','PerformanceRating','MonthlyIncome']
```

```python
column_averages = df[replaced_columns].apply(pd.to_numeric, errors='coerce').mean()
```

```python
for column in replaced_columns:
    df[column] = df[column].fillna(column_averages[column])
```

```python
df.to_csv('updated_dataset.csv', index=False)
```

# Visualisation

```python
# ATTRITION VS AGE

ax = sns.barplot(x="Attrition", y = "Age", data = df)
plt.title("Bar plot of attrition versus age")
```

```python
# we can see from the bar plot that the attrition rate is high the age is below 35.
```

```python
attrition_marital_counts = df.groupby(['Attrition', 'MaritalStatus']).size().reset_index(name='count')
plt.figure(figsize=(8, 6))
plt.title('Attrition vs. Marital Status')
sns.set_palette('Set2')  # Set color palette
sns.set_style('whitegrid')  # Set style
plt.pie(attrition_marital_counts['count'], labels=attrition_marital_counts['Attrition'] + ' - ' + attrition_marital_counts['MaritalStatus'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# The attrition is higher amongst the single employee( not married). 
```

```python
# count plot of attrition
ex = sns.countplot(  y = "Attrition", data = df)
plt.title("Count Plot of attrition")
plt.xlabel("Number of Attrition")
plt.ylabel("Attrition")
```

```python
# We can see from the above count plot the number of employee left the organisation is approximately 800 over from the couple of years.
```

```python
# countplot of worklife balance
fx = sns.countplot(x = 'WorkLifeBalance', data = df)
plt.title("Count plot of Work life balance")
plt.xlabel("Rating of work life balance")
plt.ylabel("Number of employees")
```

```python
# The count plot shows the most of the most of the employees are Satisfied with the work life balance. Less than 500 employee
# is very dissatified.
# This could be the reason for some of the employees for leaving the organisation.
```

```python
# percenthikesalary

gx = sns.histplot(x = 'PercentSalaryHike', data = df)
plt.title("Histogram of increased salary hike")
plt.xlabel("% increased in salary")
plt.ylabel("Count")
```

```python
# approximately 50 employees have 24.5% hike in salary over the couple of years.
# Most of the employees got increase in salary in between 11.5% to 12.5%
```

```python
# relation btw work at company v/s salary hike
xy = df.groupby(['YearsAtCompany','PercentSalaryHike']).size().reset_index(name='count')
gx = sns.scatterplot(x = 'PercentSalaryHike', y = 'YearsAtCompany', data = xy)
plt.title("% increased salary v/s Years at company ")
plt.xlabel("% increased in salary")
plt.ylabel("Working Years At company")
```

```python
# From the scatter plot we can depict that the employee who worked for 40 years in the organisation got 13% increase in salary.
```

```python
hx = sns.histplot(x = 'EnvironmentSatisfaction', data = df)
plt.title("Histogram of Environment Satisfaction")
plt.xlabel("Rating of Environment Satisfaction")
plt.ylabel("Frequency")
```

```python
# years since last promotion
ix = sns.countplot(x = 'YearsSinceLastPromotion', data = df)
plt.title("Bar plot of Years of since promotion")
plt.xlabel("Years since last promotion")
plt.ylabel("number of employees")
```
